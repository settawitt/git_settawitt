```python
## สุ่มเลือกข้อมูลมาเพียง 1% และใช้ pd อ่านมาเป็น pandas dataframe คือ raw_pd

f = "2008.csv"
num_lines = sum(1 for l in open(f))
size = int(num_lines / 100)
import random
skip_idx = random.sample(range(1, num_lines), num_lines - size)
raw_pd = pd.read_csv(f, skiprows=skip_idx)
```

```python
### พิสูจน์คำตอบข้อ 7

colQ7_list = [
    'CarrierDelay', 'WeatherDelay',
    'NASDelay', 'SecurityDelay', 'LateAircraftDelay'
]

for check in colQ7_list:
  display(raw_pd[raw_pd['ArrDelay']<15][check].info())
```

```python
### ครูเอ้ รู้ได้งัยว่า ต้องเป็นตัวเลข 15

raw_pd[
       (raw_pd['CarrierDelay'].isnull())\
       &(raw_pd['WeatherDelay'].isnull())\
       &(raw_pd['NASDelay'].isnull())\
       &(raw_pd['SecurityDelay'].isnull())\
       &(raw_pd['LateAircraftDelay'].isnull())\
       ].describe().transpose()
```

```python
### ครูเอ้ รู้ได้งัยว่า ต้องเป็นตัวเลข 15

raw_pd[
       (~(raw_pd['CarrierDelay'].isnull()))\
       &(~(raw_pd['WeatherDelay'].isnull()))\
       &(~(raw_pd['NASDelay'].isnull()))\
       &(~(raw_pd['SecurityDelay'].isnull()))\
       &(~(raw_pd['LateAircraftDelay'].isnull()))\
       ].describe().transpose()
```
