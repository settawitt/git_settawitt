```python
import pandas as pd 
```

```python
f = "./data/2008.csv"
```

```python
## สุ่มเลือกข้อมูลมาเพียง 1% และใช้ pd อ่านมาเป็น pandas dataframe คือ raw_pd


num_lines = sum(1 for l in open(f))
size = int(num_lines / 100)
import random
skip_idx = random.sample(range(1, num_lines), num_lines - size)
raw_pd = pd.read_csv(f, skiprows=skip_idx,header=None)
raw_pd
```

```python
# 5.ทำ subdata โดยให้ได้ dataframe ชิ้นใหม่ ชื่อ subdata_pd ซึ่งประกอบด้วย col ชื่อ ArrDelay, TaxiIn, TaxiOut, CarrierDelay,
# WeatherDelay,NASDelay,SecurityDelay,LateAircraftDelay, Year, Month, 
# DayofMonth, DayOfWeek, UniqueCarrier, DepDelay, Origin, Dest, Distance และ มี row ครบทุก row เท่ากับ raw_pd เดิม
raw_pd
```

```python
raw_pd.drop(raw_pd.iloc[0:29,21:29], inplace=True, axis=1)
raw_pd.drop(raw_pd.iloc[0:29,3:4], inplace=True, axis=1)
raw_pd.head
```

```python


raw_pd['Date'] = raw_pd[raw_pd.columns[:3]].apply(
    lambda x: '/'.join(x.dropna().astype(str)),
    axis=1
)
raw_pd.insert(11,"Year" , raw_pd[0], True)
raw_pd.insert(12,"Month" , raw_pd[1], True)
raw_pd.insert(13,"Day" , raw_pd[2], True)
raw_pd
```

```python
raw_pd[0] = raw_pd["Date"]
raw_pd.drop(raw_pd.iloc[0:23,20:23], inplace=True, axis=1)
raw_pd.drop(raw_pd.iloc[0:23,1:3], inplace=True, axis=1)
raw_pd = raw_pd.drop(columns=['Date'])
raw_pd
```

```python
raw_pd.rename(columns={'0':'Date',4:'ArrDelay',5:'TaxiIn',6:'TaxiOut',7:'CarrierDelay',
8:'WeatherDelay',9:'NASDelay',10:'SecurityDelay',11:'LateAircraftDelay','Year':'Year','Month':'Month', 
'Day':'DayofMonth',12: 'DayOfWeek',13:'UniqueCarrier',14:'DepDelay',15:'Origin',16:'Dest',17:'Distance'}, inplace=True)

raw_pd
```

```python
#6. สร้าง nonArrDelay_pd ที่มี Col ตาม subdata_pd แต่มี row เฉพาะที่ ArrDelay น้อยกว่า 15
nonArrDelay = raw_pd['ArrDelay'] = raw_pd['ArrDelay']/100
nonArrDelay = raw_pd['ArrDelay'].apply(lambda x: None if x >= 15 else x) 
nonArrDelay
```
